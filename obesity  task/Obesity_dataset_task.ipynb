Task:
-	Perform a simple linear regression to predict height using Age as the predictor.

```python
X = classwork['Age']
Y = classwork['Height']
```

```python
X = sm.add_constant(X)
linear_prediction = sm.OLS(Y,X).fit()
```

```python
linear_prediction.summary()
```

```python
#Extract independent and dependent variable 
X = classwork['Age']
Y = classwork['Height']
 
beta_0 = 164.6385  #(Intercept)
beta_1 =  0.0937  #(Slope)

#generate the trendline using regression equation
y_pred = beta_0 + beta_1*X

#Plot the scatter chart
sns.scatterplot(x = X, y = Y)
plt.title ('Age vs Height')

#Add a trendline 
plt.plot(X, y_pred, color = 'purple', label = 'Trendline')


plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

-	Interpret the regression results: What is the relationship between age and height? How much of the variance in age is explained by height?


The regression analysis produced an intercept value of **164.6385** and a slope of **0.0937**, indicating the model's linear relationship 
between Age and Height. The omnibus value of **87.466** and its **p-value (0.000)** suggest that the errors are not normally distributed. 
The **Durbin-Watson statistic (0.691)** points to positive autocorrelation, violating regression assumptions.
The **Jarque-Bera statistic (41.95)** with a **p-value < 0.05** confirms residuals do not follow a normal distribution. 
A **skew value of 0.197** indicates a slight positive skew, meaning the tails lean slightly to the right. Despite this, 
the **kurtosis value of 2.315**, close to 3, suggests residuals are near-normal.
Lastly, the condition number (**121**) indicates potential multicollinearity, 
suggesting a strong correlation between Age and Height that could impact the stability and interpretation of regression coefficients.
    

The regression shows a very weak link between **Age** and **Height**, with **height explaining only 1.3% of age's variance** (R-squared = 0.013). 
The positive slope (0.0937) suggests height slightly increases with age, but the relationship is minimal. 
While the model is statistically significant (F-statistic = 21.74, p-value < 0.0001), the low R-squared means height is not a strong predictor of age.

## Part 6: Multiple Linear Regression
Task:
-	Perform a multiple linear regression to predict overweight_obese_family using age, height, frequency of consuming vegetable, number of main meals daily, liquid intake daily, physical exercise, and schedule dedicated to technology

```python
L = classwork[['Age', 'Height', 'Physical_Excercise', 'Number_of_Main_Meals_Daily', 'Frequency_of_Consuming_Vegetables', 'Liquid_Intake_Daily', 'Schedule_Dedicated_to_Technology']]
M = classwork['Overweight_Obese_Family']
```

```python
L = sm.add_constant(L)
multiple_linear_prediction = sm.OLS(M,L).fit()
```

```python
multiple_linear_prediction.summary()
```

```python
classwork['predicted_Overweight_Obese_Family'] = multiple_linear_prediction.predict(L)
classwork[['Overweight_Obese_Family', 'predicted_Overweight_Obese_Family']].head()
```

-	Interpret the regression results: Which predictors are statistically significant? 

The multiple linear regression model to predict **Overweight_Obese_Family** shows the following insights:

1. **R-squared (0.108)** indicates that **11% of the variance** in **Overweight_Obese_Family** is explained by the 
predictors (Age, Height, Physical Exercise, etc.).
2. **Height**, **Physical Exercise**, **Number of Main Meals Daily**, **Frequency of Consuming Vegetables**, **Liquid Intake Daily**, 
and **Schedule Dedicated to Technology** are statistically significant predictors, with p-values < 0.05.
3. **Age** is not a significant predictor (p-value = 0.835).
4. The **Durbin-Watson value (1.978)** suggests no major autocorrelation issues.

In summary, while the model explains a small portion of the variance, several factors like 
**Height**, **Physical Exercise**, and **Vegetable Consumption** are important contributors to predicting **Overweight_Obese_Family**.

From the regression results, to determine which predictors are statistically significant, 
you should look at the p-values for each coefficient. If a predictorâ€™s p-value is less than 0.05, it is typically considered statistically significant. 

The regression results show that several predictors are statistically significant:

1. **Height** (p-value = 0.000), **Physical Exercise** (p-value = 0.001), **Number of Main Meals Daily** (p-value = 0.000), 
**Frequency of Consuming Vegetables** (p-value = 0.000), **Liquid Intake Daily** (p-value = 0.002), and 
**Schedule Dedicated to Technology** (p-value = 0.010) are all significant predictors of **Overweight_Obese_Family**.

2. **Age** (p-value = 0.835) is **not significant** in predicting **Overweight_Obese_Family**. 

These predictors show meaningful relationships with the outcome, except for age.
